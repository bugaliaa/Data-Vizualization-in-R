# Loading Packages

```r
library(tidyverse)
```

# Reading CSV File

```r
df<-read.csv("StudentsPerformance.csv")
```

# Display Top and Bottom few rows

```r
head(df,n=5);
tail(df,n=5);
```

# Statistical summary of the data

```r
summary(df);
```

# Correlation between quantitative variables

```r
df_quantitative<-df[,6:8];
cor(df_quantitative);
```

So, we can see that if a person is good in one subject, he will be good in other too. This is true especially for reading and writing.

# Now plotting scatter plots 

## Between math scores and reading scores

```r
plot(df_quantitative$math.score, df_quantitative$reading.score,main="Scatter plot for math scores and reading scores",
     xlab="Math scores",ylab="Reading Scores");
```

## Between math scores and writing scores

```r
plot(df_quantitative$math.score, df_quantitative$writing.score,main="Scatter plot for math scores and writing scores",
     xlab="Math scores",ylab="Writing Scores");
```

## Between reading scores and writing scores

```r
plot(df_quantitative$writing.score, df_quantitative$reading.score,main="Scatter plot for writing scores and reading scores",
     xlab="Writing scores",ylab="Reading Scores");
```

So, according to the dataset, if a person is good in one subject he is generally better in other subjects

# Box plots to visualize any outliers in the data

```r
boxplot(df_quantitative$math.score,ylab="Math Scores",main="Box plot for math scores");
```

```r
boxplot(df_quantitative$reading.score,ylab="Reading Scores",main="Box plot for Reading scores");
```

```r
boxplot(df_quantitative$writing.score,ylab="Writing Scores",main="Box plot for Writing scores");
```

So, from the data, we can see that there are certain outliers present in the dataset. But the outliers are not that far from the meat of the data, so we can say that some people do have abnormally low grades in the class.

# Density Plots

```r
dens<-density(df_quantitative$math.score);
plot(dens,xlab="Math Score");
```

```r
dens<-density(df_quantitative$reading.score);
plot(dens,xlab="Reading Score");
```

```r
dens<-density(df_quantitative$writing.score);
plot(dens,xlab="Writing Score");
```

So maximum density of students lies between 40 to 95 marks. Very intelligent students indeed.

# Categorical Plots

```r
df_qualitative<-df[,1:5];
head(df_qualitative,n=1);
```

```r
gender.freq<-table(df_qualitative$gender);
label=as.vector(names(gender.freq));
matr=as.matrix(gender.freq);
values=as.vector(matr);
#plot barplot
barplot(values,names.arg=label,main="Frequency distribution of gender",xlab="Frequency",ylab="gender");
```

More females in the dataset

```r
education_level.freq<-table(df_qualitative$parental.level.of.education);
label=as.vector(names(education_level.freq));
matr=as.matrix(education_level.freq);
values=as.vector(matr);
#plot barplot
barplot(values,names.arg=label,main="Frequency distribution of parental level of education",xlab="Frequency",ylab="Education Level",cex.names =0.5);
```

most parents are literate. Explaings why most students have decent grades

```r
ethinicity.freq<-table(df_qualitative$race.ethnicity);
label=as.vector(names(ethinicity.freq));
matr=as.matrix(ethinicity.freq);
values=as.vector(matr);
#plot barplot
barplot(values,names.arg=label,main="Frequency distribution of race ethinicity",xlab="Frequency",ylab="Ethinicity",cex.names =0.6);
```

most belong to group c and d social groups
